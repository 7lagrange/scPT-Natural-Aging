```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr
from statsmodels.stats.multitest import multipletests

import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import date
import numpy as np
import datetime

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
df1 = pd.read_csv("data/H3K9me3_gene_logFC_corr.csv")
df1.head()
```

```python
df2 = pd.read_csv("data/H3K9me3_geneTPM_corr.csv")
df2.head()
```

```python
df = pd.merge(df1, df2, how="outer", on="gene", suffixes=('_logFC', '_TPM')).sort_values("spearman_fdr_logFC")
df.head()
```

```python
df = df[['gene', 'n_points_logFC', 'spearman_r_logFC', 'spearman_p_logFC', 'spearman_fdr_logFC', 'n_points_TPM', 'spearman_r_TPM', 'spearman_p_TPM', 'spearman_fdr_TPM']]
df = df.rename(columns={
    'n_points_logFC': 'n_cell_types_logFC',
    'n_points_TPM': 'n_cell_types_TPM'
})
```

```python
df.to_csv("data/H3K9me3_gene-logFC-TPM_include-low-express-gene_corr.csv", index=False)
```
